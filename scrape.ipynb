# Reset Dataframe 

```python
tables = pd.read_html(page_source)
df = tables[0]
df.head()
```

# Remove Higher Column Headers

```python
for i in range(3):
    df.columns = df.columns.droplevel(0)
df.head()
```

```python
df = df.drop(index=[0,1])
df.reset_index(drop = True,inplace=True)
df.head()
```

# Replaced 'Principal trading partners' column name

```python
df.columns = ['Trading_Partner' if c == 'Principal trading partners' else c  for c in df.columns]
```

```python
df.columns
```

```python
for index in range(len(df)):
    for month in list(df.columns)[2:]:
        df.at[index, month] = float(df.at[index, month])
```

```python
df.Trading_Partner.unique()
```

```python
countries_list = ['United States', 'European Union5', 'China',
       'United Kingdom', 'Mexico']
df.loc[(df.Trading_Partner == 'All countries') & (df.Trade == 'Import')].values[0][2:]
val =df.loc[(df.Trading_Partner.isin(countries_list)) & (df.Trade == 'Import')].sum()[2:]
val
```

```python
countries_list = ['United States', 'European Union5', 'China',
       'United Kingdom', 'Mexico']
for trade_type in list(df.Trade.unique()):
    
    sum_all = df.loc[(df.Trading_Partner == 'All countries') & (df.Trade == str(trade_type))].values[0][2:]
    sum_others = df.loc[(df.Trading_Partner.isin(countries_list)) & (df.Trade == trade_type)].sum()[2:]
    
    sum_all_float = [float(val) for val in sum_all]
    sum_others_float = [float(val) for val in sum_others]
    difference = list(map(lambda x, y : x-y, sum_all_float, sum_others_float))
    new_row = [trade_type, "Others"] +  difference
    df.loc[len(df)] = new_row
```

```python
from datetime import datetime
data = {'data' : []}
for index,row in df.iterrows():
    for date  in df.columns[2:]:
        data['data'].append({
            'date': datetime.strptime(date, '%B %Y'),
            'trade_type': row.Trade,
            'trading_partner': row.Trading_Partner,
            'value': float(row[date])
        })
data['data'][0]
```
